## Porta

```python
def forwardMultiplyGate(x,y):
    return x * y
```

## Gradiante numérico

```python
def gradiante_numerico(funcao, x , y):
    derivada_x = 0
    derivada_y = 0
    h = 0.000001
    derivada_x = (funcao(x + h, y) - funcao(x,y)) / h
    derivada_y = (funcao(x, y + h) - funcao(x,y)) / h

    return derivada_x, derivada_y
```

```python
derivada_x, derivada_y = gradiante_numerico(forwardMultiplyGate, x, y)
print(f"Derivada em X: {derivada_x}")
print(f"Derivada em Y: {derivada_y}")
```

## Gradiente Analítico

```python
def passo_gradiente_analitico(x, y):
    derivada_x = y 
    derivada_y = x 
    
    passo = 0.01 
    
    novo_x = x + (passo * derivada_x)
    novo_y = y + (passo * derivada_y)

    return novo_x, novo_y
```

```python
derivada_x, derivada_y = gradiante_analitico(x, y)
print(f"Derivada em X: {derivada_x}")
print(f"Derivada em Y: {derivada_y}")
```
