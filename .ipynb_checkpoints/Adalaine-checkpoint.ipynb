```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import root_mean_squared_error
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```python
def Train_Adalaine(X, Y, epochs, tipo = 'classificador' ,lr=1.0):
    n_amostras = len(X)
    n_atributos = len(X[0])

    W = [0.0] * n_atributos
    b = 0.0

    for _ in range(epochs):
        for i in range(n_amostras):
            somatorioW = 0.0

            for j in range(n_atributos):
                somatorioW += X[i][j] * W[j]

            y_pred = somatorioW + b

            if(tipo == 'classificador'):
                y_pred = 1 if y_pred > 0 else 0
            
            erro = Y[i] - y_pred
        
            for j in range(n_atributos):
                W[j] += lr * erro * X[i][j]

            b += lr * erro

    return W, b
```

```python
def predict_perc(X, W, b, tipo = 'classificador'):
    n_amostras = len(X)
    n_atributos = len(X[0])
    Y = [0.0] * n_amostras

    for i in range(n_amostras):
        somatorioW = 0.0

        # soma ponderada
        for j in range(n_atributos):
            somatorioW += X[i][j] * W[j]

        y_pred = somatorioW + b
        
        if(tipo == 'classificador'):
            y_pred = 1 if y_pred > 0 else 0
        
        Y[i] = y_pred

    return Y
```

```python
###REGRESSOR
```

```python
df = pd.read_csv('notas.csv')

print(df.shape)
df.head(10)
```

```python
x = df[['prova1', 'prova2', 'prova3']].values
y = df['final'].values.reshape(-1, 1)

print(x.shape, y.shape)
```

```python
minmax = MinMaxScaler(feature_range=(-1,1))
x = minmax.fit_transform(x.astype(np.float64))
```

```python
W , b = Train_Adalaine(x, y, 150,'regressor' ,0.1)
```

```python
y_pred = predict_perc(x, W, b, 'regressor')
```

```python
y_pred = np.array(y_pred)
```

```python
y_pred
```

```python
root_mean_squared_error(y, y_pred)
```

```python
###CLASSIFICADOR
```

```python
x, y = make_blobs(n_samples=100, n_features=2, centers=2, random_state=1234)

print(x.shape, y.shape)
plt.scatter(x[:,0], x[:,1], c=y.ravel(), cmap='bwr')
```

```python
W , b = Train_Adalaine(x, y, 150,'classificador' ,0.1)
```

```python
y_pred = predict_perc(x, W, b, 'classificador')
```

```python
y_pred = np.array(y_pred)
```

```python
print(x.shape, y_pred.shape)
plt.scatter(x[:,0], x[:,1], c=y_pred.ravel(), cmap='bwr')
```
