| Nome        | Equação                                                                 | Derivada                              |
|------------|-------------------------------------------------------------------------|---------------------------------------|
| Linear     | f(x) = x                                                                 | f'(x) = 1                             |
| Sigmoide   | f(x) = 1 / (1 + e^(-x))                                                   | f'(x) = f(x)(1 − f(x))                |
| Tanh       | f(x) = tanh(x) = (2 / (1 + e^(-2x))) − 1                                 | f'(x) = 1 − f(x)²                     |
| ReLU       | f(x) = { 0, x < 0 ; x, x ≥ 0 }                                            | f'(x) = { 0, x < 0 ; 1, x ≥ 0 }        |
| Leaky ReLU | f(x) = { αx, x < 0 ; x, x ≥ 0 }                                           | f'(x) = { α, x < 0 ; 1, x ≥ 0 }        |
| ELU        | f(x) = { α(e^x − 1), x < 0 ; x, x ≥ 0 }                                   | f'(x) = { f(x) + α, x < 0 ; 1, x ≥ 0 } |

# Função Linear

$$y=x$$

$$y^\prime = 1$$

```python
def linear(x, derivative = False):
    return np.ones_like(x) if derivative else x
```

# Função Sigmoid

$$y = \frac{1}{1+e^{-x}}$$

$$y^\prime = y(1-y)$$

```python
def sigmoid(x, derivative=False):
    if derivative:
        y = sigmoide(x)
        return y * (1 - y)
    else:
        return 1 / (1 + np.exp(-x))
```

# Tangente Hiperbólica

$$y = \frac{e^x - e^{-x}}{e^x+e^{-x}}$$

$$y^\prime = 1 - y^2$$

```python
def tanh(x, derivative = False):
    if derivative:
        y = tanh(x)
        return 1 - (y ** 2)
    return (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))
```

# ReLU

$$y = max(0, x)$$

$$\frac{\partial y}{\partial x} = \begin{cases}0 &,\  x \leq 0\\1 &,\ x > 0\end{cases}$$


```python
def relu(x, derivative=False):
    if derivative: return np.where(x <= 0, 0, 1)
    return np.maximum(0, x)
```

# Leaky ReLU 

$$y = \begin{cases}\alpha x &,\ x \leq 0\\x &,\ x > 0\end{cases}$$

$$\frac{\partial y}{\partial x} = \begin{cases}\alpha &,\  x \leq 0\\1 &,\  x > 0\end{cases}$$

```python
def leaky_relu(x, derivative=False):
    alpha = 0.1
    if derivative: return np.where(x <= 0, alpha, 1)
    return np.where(x <= 0, x * alpha, x)
```

# Exponential Linear Unit (eLU) 

$$y = \begin{cases}\alpha(e^x -1) &,\ x \leq 0\\x &,\ x > 0\end{cases}$$

$$\frac{\partial y}{\partial x} = \begin{cases}y + \alpha &,\  x \leq 0\\1 &,\  x > 0\end{cases}$$

```python
def elu(x, derivative=False):
    alpha = 0.1
    if derivative:
        y = elu(x)
        return np.where(x <= 0, y + alpha, 1)
    return np.where(x <= 0, (np.exp(x) - 1) * alpha, x)
```
